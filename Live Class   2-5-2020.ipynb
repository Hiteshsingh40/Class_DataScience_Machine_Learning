# Merge , Conatenation ,join and append opertion

```python
import pandas as pd
```

create a first DataFrame

```python
raw_data={'subject_id':[1,2,3,4,5], 
          'first_name':['alex','amy','allen','alice','ayoung'], 
          'last_name':['anderson','ackerman','ali','aoni','atiches']}

df_a=pd.DataFrame(raw_data,columns=['subject_id','first_name','last_name'])
df_a
```

create a second DataFrame

```python
raw_data={'subject_id':[4,5,6,7,8],
         'first_name':['billy','brian','bran','bryce','betty'],
         'last_name':['bonder','black','balwner','brice','btisan']}

df_b=pd.DataFrame(raw_data,columns=['subject_id','first_name','last_name'])
df_b
```

create a third DataFrame

```python
raw_data={'subject_id':[1,2,3,4,5,7,8,9,10,11],
         'test_id':[51,15,15,61,16,14,15,1,61,16]}

df_n=pd.DataFrame(raw_data,columns=['subject_id','test_id'])
df_n
```

concatenation basically glues the DataFrame together concat the two DataFrame along rows by defalt axix=0 that is rows

```python
df_new=pd.concat([df_a,df_b],ignore_index=True)
df_new
```

Join the two DataFrame along columns

```python
pd.concat([df_a,df_b],axis=1)
```

# Merge two DataFrame

```python
pd.merge(df_new,df_n)
```

merge two DataFrame along the subject_id value

```python
pd.merge(df_new,df_n,on='subject_id')
```

```python
# merge with outer join

pd.merge(df_a,df_b,on='subject_id',how='outer')
```

```python
# merge with inner join

pd.merge(df_a,df_b,on='subject_id',how='inner')
```

```python
import pandas as pd
df1=pd.DataFrame({'HPI':[80,90,70,60],'INT_Rate':[2,1,2,3],'IND_GDP':[50,45,45,67]},index=[2001,2002,2003,2004])

df2=pd.DataFrame({'HPI':[80,90,70,60],'INT-Rate':[2,1,2,3],'IND_GDP':[50,45,45,67]},index=[2005,2006,2007,2008])

print(df1)
print(df2)

merged=pd.merge(df1,df2)
print(merged)
```

```python
# join opertor works join on index
df1=pd.DataFrame({'INT_Rate':[2,1,2,3],'IND_GDP':[50,45,67]},index=[2001,2002,2003,2004])
df2=pd.DataFrame({'Low_Tier_HPI':[50,45,67,34],'unemployment':[1,3,5,6]},index=[2001,2002,2003,2004])

joined=df1.join(df2)
print(joined)
```

# SET INDEX

```python
import pandas as pd
df=pd.DataFrame ({'Day':[1,2,3,4],'Visitors':[200,100,230,300],'Bounce_rate':[20,45,60,10]})

print(df)
df.set_index('Day',inplace=True)
print(df)
```

change the column Headers

```python
import pandas as pd
df=pd.DataFrame({'Day':[1,2,3,4],'Visitors':[200,100,230,300],'Bounce_Rate':{20,45,60,10}})
df=df.rename(columns={'Visitors':'Users'})
print(df)
```
